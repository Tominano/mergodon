```python
df = pd.read_excel('EODExample.xlsm', parse_dates=True) # read the downloaded data
```

```python
df = df.drop(df.columns[[0,1,7,8]], axis=1) # drop the column what we do not need
```

```python
df = df.rename(columns=df.iloc[0]).drop(df.index[0]) # add the expected name to the column
```

```python
df['Weakly maximum'] = np.nan # make the new expected column
df['Weakly minimum'] = np.nan
df['Monthly maximum'] = np.nan
df['Monthly minimum'] = np.nan
df['Quarterly maximum'] = np.nan
df['Quarterly minimum'] = np.nan
```

```python
df = df[0:-1] # I found one unnecessary row
```

```python
df = df.set_index('Date') # set the index column
```

```python
m = pd.date_range(df.index[0], df.index[-1], freq = 'M').to_series() # I get the end of the months
```

```python
s = pd.date_range(df.index[0], df.index[-1], freq = 'D').to_series() # I get end of the weeks
```

```python
month_index = ([0])
for i in range(0,len(m)):
    for j in range(0,len(df)):
        if m[i] == df.index[j]: # I make here the index of the rows of the end of the months
            month_index.append(j)
```

```python
df2 = df.copy() # make anothe table for work
```

```python
df2['ind'] = s.dt.dayofweek # set the days as indexes
```

```python
df2['num'] = (np.linspace(1,len(df),len(df))) #make an index collumn
```

```python
df2 = df2.set_index('num') # set the num column for index column
```

```python
monday_index = ([0])
for i in range(1,len(df2)):
    if (df2['ind'][i] - df2['ind'][i+1]) > 1: # Find all Monday of the data
        monday_index.append(i+1)
```

```python
weakly_max = ([])
weakly_min = ([])
for i in range(0,(int(len(monday_index)))-1):
    weakly_max.append(max((df['High'][monday_index[i]:monday_index[i+1]]))) # calculate the weakly min - max
    weakly_min.append(min((df['Low'][monday_index[i]:monday_index[i+1]])))
```

```python
monthly_max = ([])
monthly_min = ([])
for i in range(0,(int(len(month_index)))-1):
    monthly_max.append(max((df['High'][month_index[i]:month_index[i+1]])))
    monthly_min.append(min((df['Low'][month_index[i]:month_index[i+1]]))) # calculate the monthly min - max
```

```python
for i in range(len(weakly_max)):
    df['Weakly maximum'][monday_index[i+1]-1] = weakly_max[i]
    df['Weakly minimum'][monday_index[i+1]-1] = weakly_min[i] # put into the DF in the right place
```

```python
for i in range(len(monthly_max)-1):
    df['Monthly maximum'][month_index[i+1]+1] = monthly_max[i]
    df['Monthly minimum'][month_index[i+1]+1] = monthly_min[i]# put into the DF in the right place
```

```python
dates = ['2014-01-02', '2014-04-01', '2014-07-01', '2014-10-01','2015-01-01', '2015-04-01', '2015-07-01', '2015-10-01']
df3 = pd.DataFrame({'StartDate': dates})
df3['StartDate'] = pd.to_datetime(df3['StartDate'])  #made and set the quarters
df3['quarter'] = df3['StartDate'].dt.to_period('Q')
```

```python
q_index = ([])
for i in range(0,len(df3)):
    for j in range(0,len(df)):
        if df3['StartDate'][i] == df.index[j]: # get the index of the rows of end of the quarters 
            q_index.append(j)
```

```python
q_max = ([])
q_min = ([])
for i in range(0,(int(len(q_index)))-1):
    q_max.append(max((df['High'][q_index[i]:q_index[i+1]]))) # calculate the min-max
    q_min.append(min((df['Low'][q_index[i]:q_index[i+1]])))
```

```python
for i in range(len(q_max)-1):
    df['Quarterly maximum'][q_index[i+1]] = q_max[i]
    df['Quarterly minimum'][q_index[i+1]] = q_min[i]# put into the DF in the right place
```
